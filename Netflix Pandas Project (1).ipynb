### 3. What is the best time to launch a TV show?

#### Month wise

```python
# To analyze the best launch time, I first filtered TV shows from the dataset using
tv_shows_df = df[df['type'] == 'TV Show']
```

```python
# Then, I converted the date_added column to datetime format to handle missing or invalid values safely
tv_shows_df['date_added'] = pd.to_datetime(tv_shows_df['date_added'], errors='coerce')
```

```python
# viewing the new data frame
tv_shows_df
```

```python
''' From this, I extracted:
Month using .dt.month_name()
Weekday using .dt.day_name() '''

tv_shows_df['month'] = tv_shows_df['date_added'].dt.month_name()
tv_shows_df['weekday'] = tv_shows_df['date_added'].dt.day_name()
```

```python
# viewing the top new data frame
tv_shows_df.head()
```

```python
# Month-wise counts
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

# Next, I counted the number of shows added in each month to find patterns
tv_month_counts = tv_shows_df['month'].value_counts().reindex(month_order)
```

```python
# viewing the series
tv_month_counts
```

```python
#Finally, I visualized the results to identify when most shows were launched by month wise.
#Plot


plt.figure(figsize=(12,6))
sns.barplot(x=tv_month_counts.index, y=tv_month_counts.values, palette='crest')
plt.title('TV Shows Added to Netflix by Month')
plt.xlabel('Month')
plt.ylabel('Number of TV Shows Added')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Observations** :
<p>December Peak: The highest number of TV shows are added in December, suggesting a strong year-end content push.</p>

<p>Consistent Additions: Most months (except November) maintain a fairly balanced level of new additions, ranging between 10â€“14 shows.</p>

<p>November Dip: Significant drop in November, which could be due to:</p>

<p>Strategic withholding before a big December release.</p>

<p>Seasonal content planning or production delays.</p>

**Insights** :
<p>Netflix intensifies content drops in December, possibly targeting holiday binge-watchers and year-end subscriptions.</p>

<p>November is a low-activity month, which could be optimized.</p>

<p>The rest of the year shows uniformity in content releases, indicating a stable content release strategy.</p>

### 4. Heatmap of Top Genres by Country

```python


# Extract one genre per row
df_genre = df[['country', 'listed_in']].copy()
df_genre['listed_in'] = df_genre['listed_in'].str.split(',')
df_genre = df_genre.explode('listed_in')
df_genre['listed_in'] = df_genre['listed_in'].str.strip()

# Get top 10 countries and top 10 genres
top_countries = df['country'].value_counts().head(10).index
top_genres = df_genre['listed_in'].value_counts().head(10).index

# Filter
df_heat = df_genre[df_genre['country'].isin(top_countries) & df_genre['listed_in'].isin(top_genres)]

# Create pivot table
heatmap_data = df_heat.pivot_table(index='country', columns='listed_in', aggfunc='size', fill_value=0)

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap of Top Genres by Country')
plt.xlabel('Genre')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

### 4. Trend of increase in Directors over time

```python
# Filter DataFrames for Movies and TV Shows
movies_df = df[df['type'] == 'Movie'][['release_year', 'director']].dropna()
tv_df = df[df['type'] == 'TV Show'][['release_year', 'director']].dropna()

# Define bins based on min and max years in each type
movie_bins = range(int(movies_df['release_year'].min()), int(movies_df['release_year'].max()) + 5, 5)
tv_bins = range(int(tv_df['release_year'].min()), int(tv_df['release_year'].max()) + 5, 5)

# Create year bins
movies_df['year_bin'] = pd.cut(movies_df['release_year'], bins=movie_bins)
tv_df['year_bin'] = pd.cut(tv_df['release_year'], bins=tv_bins)

# Group by year_bin and count unique directors
movie_directors_per_bin = movies_df.groupby('year_bin')['director'].nunique()
tv_directors_per_bin = tv_df.groupby('year_bin')['director'].nunique()
```
